```python
# Pie chart Function

def create_pie_chart(data, column, title, color_scheme='Pastel'):
    counts = data[column].value_counts()
    pie_data = counts.reset_index()
    pie_data.columns = [column, 'Count']
    tools.display_dataframe_to_user(f"{title} Data", pie_data)


    fig = px.pie(
        names=pie_data[column],
        values=pie_data['Count'],
        title=title,
        hole=0.4,
        color_discrete_sequence=getattr(px.colors.qualitative, color_scheme)
    )
    fig.update_traces(textinfo='percent+label')
    fig.update_layout(
        title_font_size=20,
        title_x=0.5,
        margin=dict(t=50, b=50),
        legend=dict(orientation="h", yanchor="bottom", y=-0.2, xanchor="center", x=0.5)
    )
    fig.show()


```

```python
# sunburst charts Function
def create_sunburst_chart(data, path, title, color='Pastel'):
    sunburst_data = data.groupby(path).size().reset_index(name='Count')
    tools; tools.display_dataframe_to_user(f"{title} Data", sunburst_data)

    fig = px.sunburst(
        sunburst_data,
        path=path,
        values='Count',
        title=title,
        color=path[0],
        color_discrete_sequence=getattr(px.colors.qualitative, color),
        maxdepth=2
    )
    fig.update_traces(textinfo='label+percent entry')
    fig.update_layout(title_font_size=22, title_x=0.5, margin=dict(t=50, b=50))
    fig.show()

```

```python
# Pivot table Function
def create_pivot_table(data, index, columns, values, aggfunc='mean', display_name='Pivot Table'):
    pivot_df = pd.pivot_table(data, index=index, columns=columns, values=values, aggfunc=aggfunc)
    pivot_df_display = pivot_df.reset_index()
    tools.display_dataframe_to_user(display_name, pivot_df_display)
    return pivot_df_display
```



# **Key Insights Of Time Waste On Social Media**



1. **Platform Preferences by Age Group**:
   - **18-24 Year Olds**: Heavy engagement on **Instagram** and **TikTok**, favoring fast-paced and visually engaging content.
   - **25-34 Year Olds**: Balanced usage across **YouTube**, **Instagram**, and **Facebook**.
   - **35-49 Year Olds**: Preference for **Facebook** and **YouTube**, with more varied content consumption.
   - **50+ Year Olds**: Primarily use **YouTube** and **Facebook**, suggesting a trend toward long-form, informational, or entertainment content.

2. **Addiction and Engagement Times**:
   - **Peak Engagement**: Most users engage during the **evening hours (6:00 PM to 10:00 PM)**, often driven by boredom and habit.
   - **Late-Night Usage (Post 9:00 PM)**: High addiction levels noted during this time, potentially reflecting habit-forming behaviors.
   - **Morning and Afternoon Engagement**: Lower addiction levels seen during these periods, possibly indicating more intentional, task-based social media use.

3. **Content Category Preferences by Gender**:
   - **Male Users**: Engage most with **Gaming**, **Pranks**, and **Memes** content.
   - **Female Users**: Show a higher preference for **Life Hacks**, **Vlogs**, and **Entertainment** categories.
   - **Insight**: Males lean towards high-energy, entertainment-driven content, while females engage more with lifestyle and educational content.

4. **Device Usage for Video Consumption**:
   - **Smartphones** dominate, with **59%** of users consuming video content via mobile devices.
   - **Tablets and Computers** are used far less frequently, accounting for only **13-28%** of users, depending on device.
   - **Insight**: The clear preference for mobile devices underscores the need for optimized content for mobile experiences.

5. **Time Spent on Social Media**:
   - **Average Time Spent**: Users spend **148 to 228 minutes** daily across platforms, with **YouTube** and **Facebook** leading in total time spent.
   - **At Work vs At Home**: Users spend more time on social media at work, suggesting that workplace distractions are a significant source of social media usage.

6. **Reasons for Watching Videos by Time of Day**:
   - **Morning & Afternoon**: Mostly **entertainment** and **habitual viewing**, with fewer users during these hours.
   - **Evening & Night**: Peak engagement, driven by **boredom**, **procrastination**, and **habit**.
   - **Insight**: The motivation for video consumption changes based on the time of day, with late-night users engaging more due to habit or procrastination.

7. **Satisfaction and Productivity Correlation**:
   - **High Satisfaction**: Correlated with **low productivity loss**. Users who are more satisfied with their content consumption tend to report less time wasted.
   - **Low Satisfaction**: Corresponds to higher productivity loss, especially for users who engage heavily with **TikTok** or **YouTube** late into the night.
   - **Insight**: This highlights the importance of purposeful content that provides value, reducing the tendency towards unproductive usage.

8. **Platform Usage by Profession**:
   - **Students** are the largest user group, with **YouTube**, **TikTok**, and **Instagram** leading their social media activity.
   - **Waiting Staff and Labor Workers**: Show even distribution across **Facebook**, **Instagram**, and **TikTok**.
   - **Engineers and Drivers**: Lean more towards **Instagram** and **TikTok** for social media engagement.
   - **Insight**: Different professions show unique platform preferences, suggesting that targeted content strategies should consider occupational groups.

9. **Regional Video Category Preferences**:
   - **India** and the **United States** lead in overall video consumption, especially in categories like **Gaming**, **Entertainment**, and **Memes**.
   - **Germany** and **Brazil**: Show more balanced consumption across categories, but consume fewer videos overall.
   - **Insight**: Content creators should consider regional trends when targeting specific demographics for video consumption.

10. **Average Scroll Rate by Time of Day**:
    - **Morning Peak (8:00 AM - 10:00 AM)**: The highest scroll rates are observed during this time, indicating quick content browsing as users check updates before their day starts.
    - **Evening Surge (6:00 PM - 9:00 PM)**: Another significant peak, as users engage heavily after work, possibly seeking relaxation or entertainment.

11. **Video Watch Time by Connection Type**:
    - **Wi-Fi Users**: More likely to watch videos for extended periods (3-5 hours), suggesting that longer content is consumed in home environments or stable internet settings.
    - **Mobile Data Users**: Tend to engage with shorter videos (1-2 hours), likely due to data usage concerns.
    - **Insight**: Platforms should optimize for different connection types, with shorter, data-friendly content being more suited for mobile users.

12. **Addiction Level vs. Self-Control**:
    - **Insight**: As **addiction levels increase**, self-control tends to decline, especially for users heavily engaging with **TikTok** and **YouTube**. This suggests that users with moderate-to-high addiction levels may require reminders or controls to reduce their usage.

13. **Video Category Popularity Across Countries**:
    - **India** and **the United States** consume the most videos, particularly in categories like **Entertainment**, **Memes**, and **Gaming**.
    - **Smaller markets** like **Brazil**, **Germany**, and **Pakistan** show more even distribution across categories but consume less content overall.
    - **Insight**: Content tailored to local preferences (e.g., gaming videos for the U.S. and India) may yield higher engagement rates.

14. **Top Reasons for Watching by Device Type**:
    - **Smartphones**: Dominated by habitual viewing and entertainment, reflecting their role as the primary content consumption device.
    - **Tablets and Computers**: Show higher rates of boredom-driven video consumption, suggesting that users may turn to these devices for more passive entertainment or procrastination.

---

### **Conclusion: Global Trends in Social Media Consumption**

- **Evening Hours Dominate**: Engagement peaks in the evening, with users spending the most time on platforms like **YouTube**, **TikTok**, and **Instagram**. Content strategies should prioritize posting during these peak times to capture maximum attention.
  
- **Mobile-first Behavior**: With **smartphones** accounting for over **59%** of video consumption, platforms and creators should prioritize mobile-friendly content.

- **Generational Differences**: **Younger users** gravitate toward visually rich, fast-paced content on **TikTok** and **Instagram**, while **older users** prefer **YouTube** and **Facebook**, which offer longer, more in-depth videos.

- **Addiction and Self-Control**: Higher addiction levels correspond with **lower self-control**, and this is most prominent among **TikTok** and **YouTube** users. These platforms may benefit from features that help manage screen time, especially during late-night hours.

- **Regional Preferences Matter**: **India** and **the United States** dominate in terms of video consumption, with **Gaming** and **Memes** leading the way. Content localization is key to resonating with regional audiences.

- **Profession-Based Targeting**: Different professions exhibit distinct patterns of platform usage. **Students** and **waiting staff** spend the most time on **YouTube** and **Instagram**, while professionals like **engineers** and **drivers** prefer **Instagram** and **TikTok**.
